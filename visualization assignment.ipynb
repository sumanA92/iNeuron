Transform iris data into 3 dimensions
and plot a 3d chart with transformed dimensions and colour each data
point with specific class.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
from sklearn.decomposition import PCA

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
X = iris.loc[:, :'petal_width']
y = iris['species']
X.head()
```

```python
pca = PCA(n_components=3)
iris_pca = pca.fit_transform(X)
pca.explained_variance_ratio_
```

```python
iris_pca = pd.DataFrame(iris_pca, columns=['PC1', 'PC2', 'PC3'])
iris_pca['species'] = y
iris_pca.head()
```

```python
from plotly.offline import iplot
import plotly as py
import plotly.tools as tls
```

```python
#iris_pca.iplot(kind = 'scatter3d', x = 'PC1', y= 'PC2', z = 'PC3')
```

```python
y_dict = {}
for index, value in enumerate(y.unique()):
    y_dict[value] = index
    
iris_pca['species'].replace(y_dict, inplace=True)    
```

```python
import plotly.express as px 
import plotly.graph_objects as go 
    
fig = go.Figure(data =[go.Scatter3d(x = iris_pca['PC1'], y = iris_pca['PC2'], z = iris_pca['PC3'], 
                    mode ='markers', marker = dict(color = iris_pca['species'], colorscale ='Viridis', opacity = 0.5) )]) 
fig.show()
```
