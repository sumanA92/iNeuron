Matplotlib:This assignment is for visualization using matplotlib:

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
url="https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv"
titanic = pd.read_csv(url)
titanic.head()
```

1. Create a pie chart presenting the male/female proportion

```python
plt.pie(titanic['sex'].value_counts(), autopct='%1.1f%%', shadow=True, explode=(0.1, 0), 
        labels=titanic['sex'].value_counts().index)
plt.title('Male vs Female')
plt.show()
```

2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
sns.scatterplot(data=titanic, x='fare', y='age', hue='sex')
```

```python
fig = plt.scatter(data=titanic[titanic['sex'] == 'male'], x='fare', y='age', s='fare', color='orange', alpha=.7)
fig.axes.scatter(data=titanic[titanic['sex'] == 'female'], x='fare', y='age', s='fare', color='gray', alpha=.5)
```
